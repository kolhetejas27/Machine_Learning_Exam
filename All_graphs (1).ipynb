**Remove Ouliers form data**

```python
import numpy as np

# Create a sample dataset with outliers
data = [12, 14, 15, 15, 16, 16, 17, 18, 19, 20, 45, 50, 55, 100]

# Calculate the IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Define the lower and upper bounds for outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Remove outliers
filtered_data = [x for x in data if lower_bound <= x <= upper_bound]

print("Original data:", data)
print("Data after removing outliers:", filtered_data)
```

# **MATPLOTLIB GRAPHS**

```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
```

```python
# Line Plot
plt.figure(1)
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
# Scatter Plot
plt.figure(2)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
plt.scatter(x_scatter, y_scatter)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
# Bar Plot
plt.figure(3)
categories = ['Category A', 'Category B', 'Category C']
values = [20, 35, 40]
plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
```

```python
# Histogram
plt.figure(4)
data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=20)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
```

```python
# Pie Chart
plt.figure(5)
sizes = [15, 30, 45, 10]
labels = ['Slice 1', 'Slice 2', 'Slice 3', 'Slice 4']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
```

```python
# Box Plot
plt.figure(6)
data_box = np.random.normal(0, 1, 100)
plt.boxplot(data_box)
plt.title('Box Plot')
```

# **SEABORN PLOT**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a built-in dataset from Seaborn
tips = sns.load_dataset("tips")
```

```python
# Line Plot
plt.figure(figsize=(8, 6))
sns.lineplot(x="total_bill", y="tip", data=tips)
plt.title('Line Plot')
```

```python
# Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title('Scatter Plot')
```

```python
# Bar Plot
plt.figure(figsize=(8, 6))
sns.barplot(x="day", y="total_bill", data=tips)
plt.title('Bar Plot')
```

```python
# Create a bar plot with hue
plt.figure(figsize=(8, 6))
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title('Bar Plot with Hue')
plt.xlabel('Day of the week')
plt.ylabel('Total Bill')
plt.legend(title="Sex")

# Show the plot
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(tips["total_bill"], kde=True)
plt.title('Histogram')
```

```python
# Pair Plot
sns.set(style="ticks")
sns.pairplot(tips, hue="sex")
plt.suptitle('Pair Plot')
```

```python
# Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('Box Plot')
```

```python
# Heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = tips.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title('Heatmap')
```
